```python
DIRECTORY = ''

CATEGORIES = [ 'ACINETOBACTER',	      'Lactobacillus.delbrueckii',  'Propionibacterium_acnes',
'Bacteroides.fragilis'  ,'Lactobacillus.jehnsenii'	, 'Pseudomonas_aeruginosa',
'Bifidobacterium.spp'  , 'Lactobacillus.paracasei'	, 'Staphylococcus_aureus',
'Candida.albicans' ,     'Lactobacillus.plantarum',	 'Staphylococcus_epidermidis',
'Fusobacterium'	    ,  'Porfyromonas_gingivalis'	, 'Staphylococcus_saprophiticus']
```

```python
data = []

for category in CATEGORIES:
    path = os.path.join(DIRECTORY, category)
    for img in os.listdir(path):
        img_path = os.path.join(path, img)
        label = CATEGORIES.index(category)
        arr = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        new_arr = cv2.resize(arr, (60, 60))
        data.append([new_arr, label])
```

```python
data
```

```python
data[0][1]
```

```python

import random

random.shuffle(data)
```

```python
X = []
y = []
```

```python
for features, label in data:
    X.append(features)
    y.append(label)
```

```python
X = np.array(X)
y = np.array(y)
```

```python
X = X/255
X

X = X.reshape(-1, 60, 60, 1)
```

```python

from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
```

```python
from keras.layers import LSTM
```

```python
model = Sequential()
model.add(LSTM(60,activation='relu',kernel_initializer='he_uniform',input_shape=(60,60)))
model.add(Dense(60))
model.add(Dense(28, activation = 'softmax'))
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python

model.fit(X, y, epochs=1008, validation_split=0.1)
```
